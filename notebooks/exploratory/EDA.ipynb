```python
import pandas as pd

from sklearn.preprocessing import LabelEncoder, LabelBinarizer, PolynomialFeatures
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression


import os
```

```python
raw_data_path = os.path.join(os.pardir, os.pardir, "data", "raw", "all_stream_info.csv")
df = pd.read_csv(raw_data_path)
```

```python
# dropping columns: isShow, cultureGroup, isContracted, dau, privateLiveStream, streamerType, userID, registerTime, registerCountry

df.drop(columns=["isShow", "cultureGroup", "isContracted", "dau", "privateLiveStream", "streamerType", "userID", "registerTime", "registerCountry"], axis=1, inplace=True)
```

```python
df.set_index('liveStreamID', drop=True, inplace=True)
df.head()
```

```python
df.isna().sum()
```

```python
#df.totalViewerDuration.value_counts()
```

```python
# converting to Date Time

df['beginTime'] = pd.to_datetime(df['beginTime'])
df['endTime'] = pd.to_datetime(df['endTime'])
df['maxLiveViewerTime'] = pd.to_datetime(df['maxLiveViewerTime'])
```

```python
# Picking only hour

df['beginTime'] = df['beginTime'].apply(lambda x: x.hour)
df['endTime'] = df['endTime'].apply(lambda x: x.hour)
df['maxLiveViewerTime'] = df['maxLiveViewerTime'].apply(lambda x: x.hour)
```

```python
# Filling NaNs with forward fill method

df["totalViewerDuration"].fillna(method ='ffill', inplace = True)
df["avgViewerDuration"].fillna(method ='ffill', inplace = True)
df["avgStreamJoinDuration"].fillna(method ='ffill', inplace = True)
df["maxLiveViewerTime"].fillna(method ='ffill', inplace = True)
```

```python
# Replacing "closeBy" column 
# value definitions:
# normalEnd            0
# end by new stream    1
# disconnect           2
# Incoming call        3
# Killed               4
# Freezed              5
# Upload failed        6
# Crash                7
# Publish failed       8
# Keep alive failed    9

df['closeBy'] = df.closeBy.map({'normalEnd' : 0, 'end by new stream' : 1, 'disconnect' : 2, 'Incoming call' : 3, 'Killed' : 4,
                     'Freezed' : 5, 'Upload failed' : 6, 'Crash' : 7, 'Publish failed' : 8, 'Keep alive failed' : 9})
```

```python
# Dropping raws with NaNs in "closeBy" column
df = df.dropna()
```

# FSM: Linear Regression

```python
# Separating target from features

X = df.drop(['followIncreaseEstimated'], axis=1)
y = df.followIncreaseEstimated
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.25)
```

```python
fsm = LinearRegression()
```

```python
fsm.fit(X_train, y_train)
```

```python
fsm.score(X_train, y_train)
```

```python
fsm.score(X_test, y_test)
```
